```python
# Working with data types – Check the data types of all the columns and fix the incorrect ones
#(for ex. customer lifetime value and number of open complaints ). Hint: remove the percentage
#from the customer lifetime value and truncate it to an integer value.

drop_file["customer_lifetime_value"] = drop_file["customer_lifetime_value"].astype("str")

drop_file["customer_lifetime_value"] = drop_file["customer_lifetime_value"].str.replace("%"," ")

drop_file["customer_lifetime_value"] = drop_file["customer_lifetime_value"].astype("float")

drop_file
```

```python
customers_data=drop_file 
```

```python
# clean the number of open complaints and extract the middle number which is changing 
# between records.

def extract(clean):
        if clean == '1/0/00':
            clean = 0
        if clean == '1/2/00':
            clean = 2
        if clean == '1/1/00':
            clean = 1
        if clean == '1/3/00':
            clean = 3
        if clean == '1/5/00':
            clean = 5
        if clean == '1/4/00':
            clean = 4
        else:
            return clean
    
```

```python
customers_data["number_of_open_complaints"] = list(map(extract, customers_data["number_of_open_complaints"]))
```

```python
#Filtering data and Correcting typos – Filter the data in state and gender column to 
#standardize the texts in those columns

customers_data["st"].unique()
```

```python
def clean_st(i):
    if i == "Cali":
        return "California"
    if i == "AZ":
        return "Arizona"
    if i == "WA":
        return "Washington"
    else: 
        return i
```

```python
customers_data["st"] = list(map(clean_st, customers_data["st"]))
```

```python
def clean_gender(i):
    if i == "F":
        return "Female"
    if i == "female":
        return "Female"
    if i == "Femal":
        return "Female"
    if i == "M":
        return "Male"
    else:
        return (i)
```

```python
customers_data["gender"]=list(map(clean_gender, customers_data["gender"]))
```
