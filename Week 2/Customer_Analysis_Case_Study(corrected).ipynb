### Aggregate the data into one Data Frame using Pandas

```python
concatenate_table = pd.concat([file1, file2, file3])
concatenate_table
```

###  Removing duplicates

```python
Customers_data = concatenate_table.drop_duplicates()
```

### Delete the column customer

```python
#Customers_data = Customers_data.drop("customer", axis=1, inplace=1)

def customer_column(Customers_data) :
    Customers_data.drop(columns=["customer"], inplace=True)
    return Customers_data

customer_column(Customers_data)
```

### Check the data types of all the columns and fix the incorrect ones (for ex. customer lifetime value and number of open complaints ).

```python
Customers_data["customer_lifetime_value"] = Customers_data["customer_lifetime_value"].astype("str")

Customers_data["customer_lifetime_value"] = Customers_data["customer_lifetime_value"].str.replace("%"," ")

Customers_data["customer_lifetime_value"] = Customers_data["customer_lifetime_value"].astype("float")/100

Customers_data
```

### clean the number of open complaints and extract the middle number which is changing between records.

```python
Customers_data['number_of_open_complaints'].unique()
```

```python
Customers_data['number_of_open_complaints'] = Customers_data['number_of_open_complaints'].replace({'1/0/00':0,'1/2/00':2,'1/1/00':1,'1/3/00':3,'1/5/00':5,'1/4/00':4})
```

###  Filter the data in state and gender column to standardize the texts in those columns

```python
Customers_data['state'].unique()
```

```python
def clean_data(i):
    if i == 'AZ':
        i = 'Arizona'
    elif i == 'WA':
        i = 'Washington'
    elif i == 'Cali':
        i = 'California'
    return i

```

```python
Customers_data["state"] = list(map(clean_data, Customers_data["state"]))
```

```python
def clean_gender(i):
    if i == "F":
        i = "Female"
    if i == "female":
        i = "Female"
    if i == "Femal":
        i = "Female"
    if i == "M":
        i = "Male"
    else:
        return (i)
```

```python
Customers_data["gender"]=list(map(clean_gender, Customers_data["gender"]))
```

```python
Customers_data["gender"].head()
```

### Replacing null values – Replace missing values with means of the column (for numerical columns). Pay attention that the Income feature for instance has 0s which is equivalent to null values. (We assume here that there is no such income with 0 as it refers to missing values) Hint: numpy.nan is considered of float64 data type.

### Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central


```python
def replace(i):
    if i == "California":
        return "West Region"
    elif i == "Oregon":
        return "North West"
    elif i == "Washington":
        return "East"
    elif i == "Arizona" or "Nevada":
        return "Central"
    else:
        return i
    
```

```python
Customers_data = Customers_data.rename(columns={'state': 'region'})
Customers_data["region"] = list(map(replace, Customers_data["region"]))
Customers_data
```

### (Optional) In the column Vehicle Class, merge the two categories Luxury SUV and Luxury Car into one category named Luxury Vehicle

```python
Customers_data["vehicle_class"].unique()
```

```python
def luxury_name(i):
    if i in ['Luxury SUV', 'Luxury Car']:
        return 'Luxury Vehicle'
    else:
        return i
```

```python
Customers_data["vehicle_class"] = list(map(luxury_name, Customers_data["vehicle_class"]))

Customers_data["vehicle_class"].unique()
```
